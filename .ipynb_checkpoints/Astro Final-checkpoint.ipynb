```python
def auc_roc(model,y_test,X_test):
    # Binarize the target labels for multiclass ROC
    y_test_binarized = label_binarize(y_test, classes=np.unique(y_test))
    n_classes = y_test_binarized.shape[1]

    # Get the predicted probabilities
    y_score = model.predict_proba(X_test)

    # Plot ROC curve for each class
    plt.figure(figsize=(8,6))
    for i in range(n_classes):
        fpr, tpr, _ = roc_curve(y_test_binarized[:, i], y_score[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f'Class {i} (AUC = {roc_auc:.2f})')

    # Plot diagonal for random guessing
    plt.plot([0, 1], [0, 1], 'k--', label='Random Guessing')

    plt.title('ROC Curve')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    # plt.grid()

    # Print overall AUC score
    model_auc = roc_auc_score(y_test_binarized, y_score, average='macro')
    print(f"Overall AUC Score: {model_auc:.2f}")
```

# Random Search CV on Decision Tree

Using Random Search CV due to following reasons:
- Huge dataset
- Limited Computational power
- Limited time

```python
dt.get_params()
```

```python
dt_param_dict = {
    'max_depth': [None, 4, 6, 7, 8],       # Tree depth
    'min_samples_split': [2, 5, 10],           # Minimum samples to split an internal node
    'min_samples_leaf': [1, 2, 4],             # Minimum samples per leaf
    'max_features': ['sqrt', 'log2', None],    # Number of features considered for splitting
    'criterion': ['gini', 'entropy', 'log_loss'],
    'min_impurity_decrease': [0.0, 0.01, 0.1],     # Threshold to reduce impurity
}
```

```python
random_search = RandomizedSearchCV(
    estimator=dt,
    param_distributions=dt_param_dict,
    n_iter=50,
    cv=5,
    scoring='accuracy',
    verbose=2,
    random_state=42,
    n_jobs=-1
)
```

```python
import time
start = time.time()
random_search.fit(X_train,y_train)
end = time.time()
```
